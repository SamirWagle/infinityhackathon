```python
import base64

def int_to_base64(number):
    # Convert the integer to bytes
    byte_value = number.to_bytes((number.bit_length() + 7) // 8, 'big')

    # Encode the bytes to base64
    base64_value = base64.b64encode(byte_value)

    # Convert the bytes to a string
    return base64_value.decode('utf-8')

base64_result0 = int_to_base64(int(distance0))
base64_result1 = int_to_base64(int(distance1))
base64_result2 = int_to_base64(int(distance2))
print(base64_result0)
print(base64_result1)
print(base64_result2)
```

```python
import base64

def base64_to_int(base64_string):
    # Decode the base64 string to bytes
    byte_value = base64.b64decode(base64_string)

    # Convert the bytes to an integer
    int_value = int.from_bytes(byte_value, byteorder='big')

    return int_value

# Example usage:
base64_value = "IQ=="  # Base64 representation of the integer 31
int_result = base64_to_int(base64_value)
print(int_result)
```

```python
def base10_to_base128(decimal_value):
    if decimal_value == 0:
        return "0"

    base128_digits = []

    while decimal_value > 0:
        remainder = decimal_value % 128
        base128_digits.append(remainder)
        decimal_value //= 128

    base128_digits.reverse()
    return "".join(map(str, base128_digits))

# Example usage:
decimal_number = 360
base128_result = base10_to_base128(decimal_number)
print(base128_result)
```

```python
def vigenere_cipher(text, key, decrypt=False):
    key_length = len(key)
    result = []

    for i, char in enumerate(text):
        key_char = key[i % key_length]
        shift = ord(key_char) - ord('A') if key_char.isalpha() else 0

        if decrypt:
            shift = -shift

        if char.isalpha():
            shifted_char = chr(((ord(char) - ord('A') + shift) % 26) + ord('A'))
        else:
            shifted_char = char

        result.append(shifted_char)

    return ''.join(result)
```

```python
def base10_to_base64(decimal_value):
    if decimal_value == 0:
        return "0"

    base64_digits = []
    base64_characters = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+/"

    while decimal_value > 0:
        remainder = decimal_value % 64
        base64_digits.append(base64_characters[remainder])
        decimal_value //= 64

    base64_digits.reverse()
    return "".join(base64_digits)
```

```python
def base64_to_base10(base64_value):
    base64_characters = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+/"
    base64_map = {char: index for index, char in enumerate(base64_characters)}

    base64_digits = list(base64_value)
    base64_digits.reverse()

    decimal_value = 0
    multiplier = 1

    for digit in base64_digits:
        decimal_value += base64_map[digit] * multiplier
        multiplier *= 64

    return decimal_value
```
